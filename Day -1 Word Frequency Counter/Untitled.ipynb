```python
text = """
Python is a powerful programming language. Python is used in web development, data science, artificial intelligence, 
and automation. Many developers love Python because Python is easy to learn and Python has a large community. Learning Python opens doors
to many opportunites in the tech industry. Python developeers are in high demand across the world because Python is versatile and powerful."""

stopwords = {
    "the","a","an","is","in","and","to", "it","its","was","be","has","have", "of", "are","it","its","was","be","has","have","for","on",
    "at","by","from","that","this","with","as", "or"
}
```

```python
def tokenize(text):
    text = text.lower()
    text = text.replace(",","").replace(".","").replace("!","")
    text = text.replace("?","").replace(";","").replace(":","")
    words = text.split()
    return words
```

```python
def count_frequencies(words, stopwords):
    frequencies = {}
    for word in words:
        if word not in stopwords and len(word)>1:
            frequencies[word] = frequencies.get(word,0)+1
    return frequencies
```

```python
def get_top_words(frequencies, n=10):
    sorted_words = sorted(
        frequencies.items(),
        key=lambda item: item[1],
        reverse=True
    )
    return sorted_words[:n]
```

```python
def print_word_report(top_words, total_words):
    print("="*40)
    
    print("   WORD FREQUENCY REPORT   ")
    print("="*40)
    print(f"Total words (filtered):{total_words}\n")
    print(f"{'Rank':<6}{'Word':<15}{'Count':<8}")
    print("-"*40)
    for rank, (word, count) in enumerate(top_words, start=1):
        bar = " " *count
        print(f"{rank:<6}{word:<15}{count:<8}")
    print("="*40)
```

```python
def main():
    words = tokenize(text)
    frequencies = count_frequencies(words, stopwords)
    top_words = get_top_words(frequencies, n=10)
    print_word_report(top_words, len(frequencies))

#function call
main()
```

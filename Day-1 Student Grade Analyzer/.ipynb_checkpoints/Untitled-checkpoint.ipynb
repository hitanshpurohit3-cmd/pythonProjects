```python
#Student data("our database")
students = {
    "Alice":92,
    "Bob":78,
    "Charlie":55,
    "Diana":88,
    "Eve":45,
    "Frank":73,
    "Grace":95,
    "Hank":61
}
```

```python
#claculation functions
def get_average(scores):
    return sum(scores)/len(scores)

def get_highest(students):
    return max(students, key = lambda name:students[name])

def get_lowest(students):
    return min(students, key=lambda name:students[name])
```

```python
#grade distrbution
def get_grade(score):
    if score >= 90:
        return "A"
    elif score >=80:
        return "B"
    elif score >=70:
        return "C"
    elif score >=60:
        return "D"
    else:
        return "F"

def get_grade_distribution(students):
    distribution = {"A":0,"B":0,"C":0,"D":0,"F":0}
    for name, score in students.items():
        grade = get_grade(score)
        distribution[grade] +=1
    return distribution
```

```python
#The Formatted Report

def print_reports(students):
    scores=list(students.values())
    avg = get_average(scores)
    highest = get_highest(students)
    lowest = get_lowest(students)
    distribution = get_grade_distribution(students)

    print("="*40)
    print("    STUDENT GRADE REPORT    ")
    print("="*40)
    print(f"Total Students : {len(students)}")
    print(f"Class Average : {avg:.2f}")
    print(f"Highest Score: {highest}({students[highest]})")
    print(f"Lowest Score: {lowest}({students[lowest]})")
    print("\n---Grade Distribution---")
    for grade, count in distribution.items():
        bar = " "*count
        print(f"{grade}:{bar}({count})")
        print("="*40)

#function call
print_reports(students)

```

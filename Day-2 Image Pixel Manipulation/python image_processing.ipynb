```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#--1.Load Image--#
#imread returns a Numpy array of shape (height, width, 3)
#Each pixel = [R,G,B] with values 0-255 (uint8) or 0.0-1.0(float)
img = mpimg.imread('photo.jpg')

print(f"Shape: {img.shape}")
print(f"Dtype: {img.dtype}")

# Normalize to 0.0-1.0 if the image is uint8(0-255)
if img.dtype == np.uint8:
    img = img/255.0
```

```python
#--2.Grayscale (Luminosity Formula)--#
#Human eyes are most sensitive to gree, least to blue
#we take a weighted sum across the last axis (the RGB channel axis)
#img[:,:,0] = all Red values, [:,:,1] =Green,[:,:,2]=Blue
gray = (0.2126 * img[:,:,0]+ 0.7152 * img[:,:,1]+0.0722 * img[:,:,2])
#gray is now shape (height,width) - one value per pixel
```

```python
#--3.Brightness--#
# multiply every pixel by a factor >1 to brighten
#np.clip ensures no value exceeds 1.0(which would wrap/ corrupt colors)
bright = np.clip(img*1.4,0.0,1.0) #40% brighter
```

```python
#--4.Plot: 3 Panels--#
fig ,axes = plt.subplots(1,3,figsize=(15,5))

axes[0].imshow(img)
axes[0].set_title('Original',fontsize=13)
axes[0].axis('off')

axes[1].imshow(gray, cmap='gray')
axes[1].set_title('Grayscale (Luminosity)', fontsize=13)
axes[1].axis('off')

axes[2].imshow(bright)
axes[2].set_title('Brightness +40%', fontsize=13)
axes[2].axis('off')

plt.suptitle('Numpy Image Processing', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('image_processing.png',dpi=150)
plt.show()
```

```python
#--5.Pixel Stats--#
print(f"\nOriginal mean brightness: {np.mean(img):.3f}")
print(f"Brightened mean: {np.mean(bright):.3f}")
print(f"Gray pixel range: {gray.min():.3f}-{gray.max():.3f}")
```
